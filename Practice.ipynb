```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('creditcard.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data['Class'].value_counts()
```

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#clearly we can see the differnce between the fraud and legit transaction
# compare both the value for both transaction

data.groupby('Class').mean()
```

```python
#we are going to create a sample dataset containing the similar distributtin of
#normal transaction and fraudulent transaction
```

```python
#no. of fraud transaction -> 492  

legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
print(fraud.shape)
```

```python
new_data = pd.concat([legit_sample,fraud], axis= 0)
```

```python
new_data.head()
```

```python
new_data.groupby('Class').mean()
```

```python
#here point to be noted is that the mean values for earlier and newer dataset is not much
#not much different which suggest that sample taken is good


#Splitting data into Training and test data 

X = new_data.drop(columns = 'Class', axis = 1)
Y = new_data['Class']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled,Y,test_size=0.2, stratify=Y,random_state=2)
```

```python
#we are using logistic regression 

model = LogisticRegression(max_iter=1000, solver = 'lbfgs')
```

```python
#training the model using training dataset

model.fit(X_train,Y_train)
```

```python
#model evaluation
```

```python
#accuracy Score

#accuracy on trianing 

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(test_data_accuracy)
```
